```python
import urllib
import warnings
from pathlib import Path as p
from pprint import pprint

import pandas as pd
from langchain import PromptTemplate
from langchain.chains.question_answering import load_qa_chain
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA



warnings.filterwarnings("ignore")
# restart python kernal if issues with langchain import.
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
model = ChatGoogleGenerativeAI(model="gemini-pro",google_api_key=GOOGLE_API_KEY,
                             temperature=0.2,convert_system_message_to_human=True)
```

### Extract text from the PDF

```python
pdf_loader = PyPDFLoader("/content/DATASET_1.pdf")
pages = pdf_loader.load_and_split()
print(pages[3].page_content)
```

```python
len(pages)
```

### RAG Pipeline: Embedding + Gemini (LLM)

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=1000)
context = "\n\n".join(str(p.page_content) for p in pages)
texts = text_splitter.split_text(context)
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001",google_api_key=GOOGLE_API_KEY)
```

```python
vector_index = Chroma.from_texts(texts, embeddings).as_retriever(search_kwargs={"k":5})
```

```python
qa_chain = RetrievalQA.from_chain_type(
    model,
    retriever=vector_index,
    return_source_documents=True

)
```

```python
template = """Use the following pieces of context to answer the question at the end. If you don't know the answer, say " I am sorry, I don't know the answer", don't try to make up an answer. Keep the answer as concise as possible. Always say "Thanks for asking!" at the end of the answer.
{context}
Question: {question}
Helpful Answer:"""
QA_CHAIN_PROMPT = PromptTemplate.from_template(template)# Run chain
qa_chain = RetrievalQA.from_chain_type(
    model,
    retriever=vector_index,
    return_source_documents=True,
    chain_type_kwargs={"prompt": QA_CHAIN_PROMPT}
)
```

```python
question = " What is the best time to visit the suchipara waterfalls?"
result = qa_chain({"query": question})
Markdown(result["result"])
```
